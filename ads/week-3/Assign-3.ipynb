```python
import pandas as pd
import numpy as np

%xmode plain
```

```python
energy_ind = pd.ExcelFile('Energy Indicators.xls')

#read sheet
energy = energy_ind.parse('Energy')
energy.head()
```

```python
energy = pd.read_excel('Energy Indicators.xls', skiprows=17, skipfooter=38)
energy.drop(energy.columns[[0, 1]], axis=1, inplace=True)
energy.head()
```

```python
energy.describe(include='all')
```

```python
energy.rename(columns={'Unnamed: 2': 'Country', 'Petajoules':'Energy Supply', 
                       'Gigajoules': 'Energy Supply per Capita', '%':'% Renewable'}, inplace=True)
energy.head()
```

```python
energy[(energy['Energy Supply'] == '...') & (energy['Energy Supply per Capita'] == '...')]
```

```python
energy = energy[(energy['Energy Supply'] != '...') & (energy['Energy Supply per Capita'] != '...')]
energy.head()
```

```python
energy['Energy Supply'] = energy['Energy Supply'] * 1000000
energy.head()
```

```python
energy = energy.dropna(axis=0)
energy.head()
```

```python
energy = energy.replace('Bolivia (Plurinational State of)', 'Bolivia')
energy = energy.replace('United States of America20', 'United States')
energy = energy.replace('United Kingdom of Great Britain and Northern Ireland19', 'United Kingdom')
energy = energy.replace('China, Hong Kong Special Administrative Region', 'Hong Kong')
energy = energy.replace('Venezuela (Bolivarian Republic of)', 'Venezuela')
energy = energy.replace('Falkland Islands (Malvinas)', 'Falkland Islands')
energy = energy.replace('Iran (Islamic Republic of)', 'Iran')
energy = energy.replace('Micronesia (Federated States of)', 'Micronesia')
energy = energy.replace('Sint Maarten (Dutch part)', 'Sint Maarten')
#energy[energy['Country'] == 'Bolivia (Plurinational State of)'] = 'Bolivia'
#energy[energy['Country'] == 'United States of America20'] = 'United States'
#energy[energy['Country'] == 'United Kingdom of Great Britain and Northern Ireland19'] = 'United Kingdom'
#energy[energy['Country'] == 'China, Hong Kong Special Administrative Region'] = 'Hong Kong'
energy.head()
```

```python
energy['Country'] = energy['Country'].str.replace('\d+', '')
energy.head()
```
